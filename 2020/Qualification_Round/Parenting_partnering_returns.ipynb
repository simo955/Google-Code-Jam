```python
def resolve_prob(tasks_list):
    new_tasks_list = []
    organizer_string = []
    for i in range(len(tasks_list)):
        numbers = [int(s) for s in tasks_list[i].split() if s.isdigit()]
        new_tasks_list = new_tasks_list+numbers

    jamie_list = []
    cameron_list = []
    p=0
    for i in range(0, len(new_tasks_list), 2):
        p+=1
        if i == 0:
            jamie_list.append(new_tasks_list[i])
            organizer_string.append('J')
            jamie_list.append(new_tasks_list[i+1])
        else:
            for t in range(0, len(jamie_list), 2):
                print(organizer_string)
                print('T',t)
                i_task = jamie_list[t]
                f_task = jamie_list[t+1]
                i_new_task = new_tasks_list[i]
                f_new_task = new_tasks_list[i+1]
                if ((i_new_task < f_task)and(i_new_task >= i_task))or((f_new_task > i_task)and(f_new_task <= f_task)):
                    for t1 in range(0,len(cameron_list),2):
                        print('t2',t1)
                        i_task=cameron_list[t1]
                        f_task=cameron_list[t1+1]
                        if ((i_new_task < f_task)and(i_new_task>=i_task))or((f_new_task>i_task)and(f_new_task<=f_task)):
                            return ('IMPOSSIBLE')
                    if len(organizer_string)<p:
                        print("AGGIUNGO C")
                        organizer_string.append('C')
                        cameron_list.append(i_new_task)
                        cameron_list.append(f_new_task)
                        print("XXX",organizer_string)
                        break
            if len(organizer_string)<p:      
                print("AGGIUNGO J")
                organizer_string.append('J')
                jamie_list.append(i_new_task)
                jamie_list.append(f_new_task)
                print("YYYY",organizer_string)
            
    return  ''.join(organizer_string)


#Read inpunt
t = int(input()) 
for i in range(1, t + 1):
    n = int(input())
    tasks_list=[]
    for y in range(n):
        tasks_list.append(str(input()))
    
    organizer_string=resolve_prob(tasks_list)
    
    #format output
    print("Case #{}: {}".format(i, organizer_string))
```
